```python
month_dum = pd.get_dummies(df['MONTH'])
month_dum.head()
```

```python
df.plot()
```

```python
plt.figure(figsize=(6, 6))
sns.lineplot(data=df, x='STORECODE', y='BILL_AMT')
plt.title('Trends in Bill Amount over Time in a particular storecode')
plt.xlabel('Day')
plt.ylabel('Bill Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='MONTH', y='STORECODE')
plt.title('GRAPH OF SALES IN THE DIFFERENT STORES')
plt.xlabel('Month')
plt.ylabel('Storecode')
plt.show()
```
